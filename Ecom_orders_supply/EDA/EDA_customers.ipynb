```python
# перезапишем датафрейм и сразу установим тип данных для индекса как "строка"
df_c = pd.read_csv(
    '../raw_data/df_Customers.csv',
    dtype={'customer_zip_code_prefix': 'str'})

df_c.info()
```

```python
df_c.head()

# проверка показала, что в исходном файле уже отсутствовали ведущие нули
```

```python
# в Бразилии префикс индекса содержит 5 цифр. Восстановим ведущие нули
df_c['customer_zip_code_prefix'] = (
    df_c['customer_zip_code_prefix']
    .str.zfill(5)
)
df_c.head()
```

```python
# проверим отсутствующие значения
df_c.isna().sum()
```

```python
# проверим дубликаты
df_c.duplicated().sum()
```

```python
# проверим уникальные значения
df_c.nunique()

# количество по customer_id соответствует количеству строк в общем => таблица на уровне "кникальный клиент",
# т.е. каждому клиенту соответствует отдельная строка
```

```python
# найдем топ штатов по числу заказов / клиентов

state_counts = (
    df_c.groupby('customer_state', as_index=False)
    .agg(customer_count=('customer_id', 'count'))
)

state_counts['share_pct'] = (
    state_counts['customer_count']
    / state_counts['customer_count'].sum()
    * 100
)

state_counts = state_counts.sort_values('customer_count', ascending=False)
state_counts

# более 65% всех заказов приходятся на три штата: Сан-Паулу, Рио-де-Жанейро и Минас-Жерайс
# более 42% всех заказов приходится на штат Сан-Паулу
# имеем ярко выраженную концентрацию заказов в трех крупнейших штатах Юго-Востока, что типично для бразильского ecom
```

```python
# найдем топ городов по числу заказов / клиентов

city_counts = (
    df_c.groupby('customer_city', as_index=False)
    .agg(customer_count=('customer_id', 'count'))
)

city_counts['share_pct'] = (
    city_counts['customer_count']
    / city_counts['customer_count'].sum()
    * 100
)

city_counts = city_counts.sort_values('customer_count', ascending=False)
city_counts.head(10)

# основная доля заказов отгружается в город Сан-Паулу (штат Сан-Паулу) (>16%)
# вторым по количеству заказов идет Рио-де-Жанейро (штат Рио-де-Жанейро) (~7%)
# третьим по количеству заказов идет Белу-Оризонти (штат Минас-Жирайс) (~2.7%)
```
