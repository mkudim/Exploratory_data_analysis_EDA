```python
# перезапишем датафрейм и сразу установим тип данных для индекса как "строка"
df_c = pd.read_csv(
    '../raw_data/df_Customers.csv',
    dtype={'customer_zip_code_prefix': 'str'})

df_c.info()
```

```python
df_c.head()

# проверка показала, что в исходном файле уже отсутствовали ведущие нули
```

```python
# в Бразилии префикс индекса содержит 5 цифр. Восстановим ведущие нули
df_c['customer_zip_code_prefix'] = (
    df_c['customer_zip_code_prefix']
    .str.zfill(5)
)
df_c.head()
```

```python
# проверим отсутствующие значения
df_c.isna().sum()
```

```python
# проверим дубликаты
df_c.duplicated().sum()
```

```python
# проверим уникальные значения
df_c.nunique()

# количество по customer_id соответствует количеству строк в общем => таблица на уровне "кникальный клиент",
# т.е. каждому клиенту соответствует отдельная строка
```

```python
# найдем топ штатов по числу клиентов
state_counts = (
    df_c.groupby('customer_state', as_index=False)
    .agg(customer_count=('customer_id', 'count'))
)

state_counts['share_pct'] = (
    state_counts['customer_count']
    / state_counts['customer_count'].sum()
    * 100
)

state_counts = state_counts.sort_values('customer_count', ascending=False)
state_counts.head(10)

# более 65% всех заказов приходятся на три штата: Сан-Паулу, Рио-де-Жанейро и Минас-Жерайс
# более 42% всех заказов приходится на штат Сан-Паулу
# имеем ярко выраженную концентрацию заказов в трех крупнейших штатах Юго-Востока, что типично для бразильского ecom
```

```python
# найдем топ городов по числу заказов / клиентов
city_counts = (
    df_c.groupby('customer_city', as_index=False)
    .agg(customer_count=('customer_id', 'count'))
)

city_counts['share_pct'] = (
    city_counts['customer_count']
    / city_counts['customer_count'].sum()
    * 100
)

city_counts = city_counts.sort_values('customer_count', ascending=False)
city_counts.head(10)

# основная доля заказов отгружается в город Сан-Паулу (штат Сан-Паулу) (>16%)
# вторым по количеству заказов идет Рио-де-Жанейро (штат Рио-де-Жанейро) (~7%)
# третьим по количеству заказов идет Белу-Оризонти (штат Минас-Жирайс) (~2.7%)
```

```python
# проверим склько заказов у каждого клиента. Работать будет с df_o_clean, в котором удалены аномалии даты
df_o_clean = pd.read_pickle('df_o_clean.pkl')
df_o_clean.head()
```

```python
df_orcu = df_o_clean.merge(df_c, on='customer_id', how='inner')
df_orcu.head()
```

```python
df_cnt_orders = df_orcu.groupby('customer_id').agg(cnt=('order_id', 'count')).reset_index().query('cnt > 1')
df_cnt_orders.head()

# каждому клиенту соответствует только один заказ, т.е. F=1
# дальнейший анализ по F и количеству заказов не даст сегментации
```

```python
# проверим согласованность между индексами и городами /штатами
df_orcu.groupby('customer_state').agg(cnt_city=('customer_city', 'count')).reset_index()
```

```python
df_ind_state = (
    df_orcu.groupby('customer_zip_code_prefix')
    .agg(cnt_state=('customer_state', 'count'))
    .reset_index()
)
df_ind_state.describe()
```

```python
df_ind_city = (
    df_orcu.groupby('customer_zip_code_prefix')
    .agg(cnt_city=('customer_city', 'count'))
    .reset_index()
)
df_ind_city.describe()
```

```python
df_ind_state_city = (
    df_orcu.groupby('customer_zip_code_prefix')
    .agg({'customer_city': 'nunique', 'customer_state': 'nunique'})
    .reset_index()
    .query('customer_city > 1')
)
df_ind_state_city

# проерка показывает, что аномальный данных по индексам (например, один индекс три штата) у нас нет
```

```python
# проверим есть ли заказы без указания id клиента
missing_orders = df_o_clean[~df_o_clean['customer_id'].isin(df_c['customer_id'])]
print(f"Заказы без клиентов: {len(missing_orders)}")
```

```python
city_counts = (
    df_c.groupby('customer_city', as_index=False)
    .agg(customer_count=('customer_id', 'count'))
)

city_counts['share_pct'] = (
    city_counts['customer_count']
    / city_counts['customer_count'].sum()
    * 100
)

city_counts = city_counts.sort_values('customer_count', ascending=False)
city_counts.head(10)
```

```python
# сравним распределение заказов и распределение клиентов по штатам
state_distr = (
    df_orcu.groupby('customer_state', as_index=False)
    .agg({'customer_id': 'nunique', 'order_id': 'nunique'})
)

state_distr = state_distr.rename(columns={
    'customer_id': 'cnt_client',
    'order_id': 'cnt_order'
})

total_clients = state_distr['cnt_client'].sum()
total_orders = state_distr['cnt_order'].sum()

state_distr['pct_client'] = (state_distr['cnt_client'] / total_clients * 100).round(2)
state_distr['pct_order'] = (state_distr['cnt_order'] / total_orders * 100).round(2)

state_distr = state_distr.sort_values('pct_client', ascending=False)
state_distr[['customer_state', 'cnt_client', 'pct_client', 'cnt_order', 'pct_order']].head()
```

```python
# посчитаем churn для клиентов
orders_per_cust = (
    df_orcu.groupby('customer_id', as_index=False)
    .agg(cnt_orders=('order_id', 'nunique'))
)

one_order_share = (orders_per_cust['cnt_orders'] == 1).mean()
one_order_share

# Доля клиентов с только одним заказом = 100%.
# Это означает, что датасет отражает только первый заказ каждого клиента
# (или клиенты не делают повторных покупок в наблюдаемом периоде),
# поэтому классический churn‑анализ по повторным покупкам и частоте F в RFM невозможен.
# Можно анализировать только recency и monetary
# Классический когортный анализ проводится тоже не будет, потому что нет повторных покупок
```
