```python
# проверим аномальные цены
quantiles = df_oi[['price', 'shipping_charges']].quantile([0.01, 0.25, 0.5, 0.75, 0.99])
quantiles
```

```python
iqr_price = quantiles.loc[0.75, 'price'] - quantiles.loc[0.25, 'price']
iqe_shipping_charges = quantiles.loc[0.75, 'shipping_charges'] - quantiles.loc[0.25, 'shipping_charges']
```

```python
low_limit_price = quantiles.loc[0.25, 'price'] - 1.5 * iqr_price
high_limit_price = quantiles.loc[0.75, 'price'] + 1.5 * iqr_price

print(f'Нижняя граница {low_limit_price.round(2)}', end='\n')
print(f'Верхняя граница {high_limit_price.round(2)}')
```

```python
low_limit_ship_charge = quantiles.loc[0.25, 'shipping_charges'] - 1.5 * iqe_shipping_charges
high_limit_ship_charge = quantiles.loc[0.75, 'shipping_charges'] + 1.5 * iqe_shipping_charges

print(f'Нижняя граница {low_limit_ship_charge.round(2)}', end='\n')
print(f'Верхняя граница {high_limit_ship_charge.round(2)}')
```

```python
# кол-во price выше верхней границы
df_oi.query('price > @high_limit_price').shape[0]
```

```python
# кол-во shipping_charges выше верхней границы
df_oi.query('shipping_charges > @high_limit_ship_charge').shape[0]
```
