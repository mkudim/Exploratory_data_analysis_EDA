```python
# проверим аномальные цены
quantiles = df_oi[['price', 'shipping_charges']].quantile([0.01, 0.25, 0.5, 0.75, 0.99])
quantiles
```

```python
iqr_price = quantiles.loc[0.75, 'price'] - quantiles.loc[0.25, 'price']
iqe_shipping_charges = quantiles.loc[0.75, 'shipping_charges'] - quantiles.loc[0.25, 'shipping_charges']
```

```python
low_limit_price = quantiles.loc[0.25, 'price'] - 1.5 * iqr_price
high_limit_price = quantiles.loc[0.75, 'price'] + 1.5 * iqr_price

print(f'Нижняя граница {low_limit_price.round(2)}', end='\n')
print(f'Верхняя граница {high_limit_price.round(2)}')
```

```python
low_limit_ship_charge = quantiles.loc[0.25, 'shipping_charges'] - 1.5 * iqe_shipping_charges
high_limit_ship_charge = quantiles.loc[0.75, 'shipping_charges'] + 1.5 * iqe_shipping_charges

print(f'Нижняя граница {low_limit_ship_charge.round(2)}', end='\n')
print(f'Верхняя граница {high_limit_ship_charge.round(2)}')
```

```python
# кол-во price выше верхней границы
df_oi.query('price > @high_limit_price').shape[0]
```

```python
# кол-во shipping_charges выше верхней границы
df_oi.query('shipping_charges > @high_limit_ship_charge').shape[0]
```

```python
df_oipay = df_oi.merge(df_p_clean, on='order_id', how='inner')
df_oipay[['order_id', 'total_amount', 'payment_value']]
```

```python
df_oipay['is_anomaly_price'] = df_oipay['payment_value'] < df_oipay['price']
df_oipay['is_anomaly_price_shipping'] = df_oipay['payment_value'] < df_oipay['total_amount']
df_oipay[['order_id', 'is_anomaly_price', 'is_anomaly_price_shipping']]
```

```python
price_anomalies = df_oipay['is_anomaly_price'].sum()
shipping_anomalies = df_oipay['is_anomaly_price_shipping'].sum()

print(f"Аномалий цены: {price_anomalies}")
print(f"Аномалий с доставкой: {shipping_anomalies}")
```

```python
anomaly_table = pd.DataFrame({
    'Тип аномалии': ['Ценовая аномалия (price)', 'Аномалия доставки (shipping)'],
    'Количество': [price_anomalies, shipping_anomalies],
    'Процент': [
        round(price_anomalies / len(df_oipay) * 100, 2),
        round(shipping_anomalies / len(df_oipay) * 100, 2)
    ]
})

anomaly_table
```

```python
# суммы по прайсам (предположим, что доставка включена в цену товара)
sum_price_anomaly_true = df_oipay.loc[df_oipay['is_anomaly_price'], 'price'].sum()
sum_price_anomaly_false = df_oipay.loc[~df_oipay['is_anomaly_price'], 'price'].sum()

# суммы по прайсам + логистика
sum_total_anomaly_true = df_oipay.loc[df_oipay['is_anomaly_price'], 'total_amount'].sum()
sum_total_anomaly_false = df_oipay.loc[~df_oipay['is_anomaly_price'], 'total_amount'].sum()

# суммы оплат от клиентов по заказам
sum_payment_anomaly_true = df_oipay.loc[df_oipay['is_anomaly_price'], 'payment_value'].sum()
sum_payment_anomaly_false = df_oipay.loc[~df_oipay['is_anomaly_price'], 'payment_value'].sum()

print(f'Сумма аномальной стоимости заказов (доставка в цене): {sum_price_anomaly_true:,.2f}')
print(f'Сумма корректной стоимости заказов (доставка в цене): {sum_price_anomaly_false:,.2f}')
print(f'Общая сумма стоимости заказов (доставка в цене): {sum_price_anomaly_true + sum_price_anomaly_false:,.2f}')
print()
print(f'Сумма аномальной стоимости заказов (доставка отдельно): {sum_total_anomaly_true:,.2f}')
print(f'Сумма корректной стоимости заказов (доставка отдельно): {sum_total_anomaly_false:,.2f}')
print(f'Общая сумма стоимости заказов (доставка отдельно): {sum_total_anomaly_true + sum_total_anomaly_false:,.2f}')
print()
print(f'Сумма оплат клиентами: {sum_payment_anomaly_true:,.2f}')
print(f'Сумма оплат клиентами: {sum_payment_anomaly_false:,.2f}')
print(f'Общая сумма оплат клиентами: {sum_payment_anomaly_true + sum_payment_anomaly_false:,.2f}')
```

```python
df_oipay['price_diff'] = df_oipay['price'] - df_oipay['payment_value']
df_oipay['price_diff'].describe()

# Если price_diff > 0 → клиент заплатил меньше, чем цена
# Если price_diff < 0 → клиент заплатил больше, чем цена
```

```python
stats_anomaly = df_oipay.loc[df_oipay['is_anomaly_price'], 'price_diff'].describe()
stats_anomaly

# В половине «аномальных» заказов недоплата ≥ ≈208 единиц.
# В четверти — недоплата ≥ ≈527.
# Есть хвост до ≈6.7K.
```

```python
stats_ok = df_oipay.loc[~df_oipay['is_anomaly_price'], 'price_diff'].describe()
stats_ok

# Во всех этих заказах payment_value ≥ price, и price_diff ≤ 0.
# Медианная переплата ≈150.76, четверть заказов имеет переплату ≥≈308 по модулю.
# Хвост переплат до ≈7K.
```
