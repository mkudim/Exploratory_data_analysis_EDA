```python
# соберем витрину с заказами, оплатами и скрытыми скидками
fact_orders_finance  = df_oipay[['order_id', 'product_id', 'price', 'shipping_charges', 'total_amount', 'payment_value']]
fact_orders_finance
```

```python
fact_orders_finance['hidden_discount_abs'] = fact_orders_finance['total_amount'] - fact_orders_finance['payment_value']
fact_orders_finance['hidden_discount_pct '] = round(fact_orders_finance['hidden_discount_abs'] / fact_orders_finance['total_amount'] * 100)
fact_orders_finance['payment_ratio_pct'] = round(fact_orders_finance['payment_value'] / fact_orders_finance['total_amount'] * 100)
```

```python
fact_orders_finance = fact_orders_finance.drop(columns='payment_ratio ')
```

```python
fact_orders_finance.describe()
```

```python
# 1. Определяем границы бинов (в процентах)
bins = [-100, 0, 10, 20, 30, 50, 100, 1000]
labels = [
    '≤0% (нет скидки / переплата)',
    '0–10%',
    '10–20%',
    '20–30%',
    '30–50%',
    '50–100%',
    '>100% (аномалия)'
]

# 2. Создаём категориальный столбец
fact_orders_finance['hidden_discount_band'] = pd.cut(
    fact_orders_finance['hidden_discount_pct '],
    bins=bins,
    labels=labels,
    right=True,
    include_lowest=True
)

# 3. Считаем распределение по бинам
discount_dist = (
    fact_orders_finance['hidden_discount_band']
    .value_counts(dropna=False)
    .to_frame('orders_count')
    .assign(share_pct=lambda df: df['orders_count'] / df['orders_count'].sum() * 100)
    .sort_index()
)

discount_dist

# «номинал» (total_amount) существенно выше фактической оплаты
# либо платформа/маркетплейс активно субсидирует,
# либо прайс-лист сильно завышен и всегда продаётся «со скидкой»
```

```python
fact_orders_finance['discount_type'] = np.where(
    fact_orders_finance['hidden_discount_pct '] < -100,
    'overpay_or_miscalc',      # заплатили намного больше номинала / проблема расчёта
    'normal_or_discount'       # всё остальное
)
fact_orders_finance.head()
```

```python
discount_type_counts = (
    fact_orders_finance['discount_type']
    .value_counts(dropna=False)
    .to_frame('rows_count')
    .assign(share_pct=lambda df: df['rows_count'] / df['rows_count'].sum() * 100)
)

discount_type_counts
```

```python
# посмотрим есть ли зависимость скидки от типа оплаты
# пример: берём payment_type платежа с максимальным payment_value
payments_main = (
    df_p_clean
    .sort_values(['order_id', 'payment_value'], ascending=[True, False])
    .drop_duplicates(subset='order_id')  # остаётся по одному payment_type на заказ
    [['order_id', 'payment_type']]
)
payments_main
```

```python
fact_orders_finance = fact_orders_finance.merge(
    payments_main,
    on='order_id',
    how='left'
)
fact_orders_finance
```

```python
# 1. Оставляем только "нормальные" заказы (без сильной переплаты/аномалий)
df_disc = fact_orders_finance[
    fact_orders_finance['discount_type'] == 'normal_or_discount'
].copy()

# 2. Группировка по payment_type с основными метриками по скидкам
discount_by_paytype = (
    df_disc
    .groupby('payment_type')
    .agg(
        orders_count=('order_id', 'nunique'),
        avg_hidden_discount_pct=('hidden_discount_pct ', 'mean'),
        median_hidden_discount_pct=('hidden_discount_pct ', 'median'),
        avg_payment_ratio_pct=('payment_ratio_pct', 'mean')
    )
    .reset_index()
)

# 3. Доля заказов каждого payment_type среди всех "нормальных"
total_orders_disc = discount_by_paytype['orders_count'].sum()
discount_by_paytype['orders_share_pct'] = (
    discount_by_paytype['orders_count'] / total_orders_disc * 100
)

# 4. Округление для удобства чтения
cols_to_round = [
    'avg_hidden_discount_pct',
    'median_hidden_discount_pct',
    'avg_payment_ratio_pct',
    'orders_share_pct'
]
discount_by_paytype[cols_to_round] = discount_by_paytype[cols_to_round].round(2)

discount_by_paytype

# размер скидки не зависит от способа оплаты
```
