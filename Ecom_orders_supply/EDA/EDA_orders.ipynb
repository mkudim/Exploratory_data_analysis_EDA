```python
# переименуем колонки для удобства
df_o = df_o.rename(columns={
                   'order_purchase_timestamp': 'purchased_date',
                   'order_approved_at': 'approved_at_date',
                   'order_delivered_timestamp': 'delivered_date',
                   'order_estimated_delivery_date': 'estimated_delivery_date'}
                   )

df_o.columns
```

```python
# переопределим тип у колонок с датами
date_cols = [
    'purchased_date',
    'approved_at_date',
    'delivered_date',
    'estimated_delivery_date'
]

df_o[date_cols] = df_o[date_cols].apply(pd.to_datetime)

df_o.info()
```

```python
# проверим отсутствующие значения
df_o.isna().sum()

# в датасете есть отсутствующие значения, и они достаточно логичны:
# 9 заказов еще ожидает подтверждения
# 1889 заказов еще находятся в пути до клиента
```

```python
# проверим дубликаты
df_o.duplicated().sum()
```

```python
# проверим уникальные значения
df_o.nunique()
```

```python
df_o['order_status'].unique()
```

```python
# переведем данные в столбце 'order_status' в тип данных 'категория'
df_o['order_status'] = df_o['order_status'].astype('category')
df_o.info()
```

```python
df_o.describe()
```

```python
# посмотрим распределение по статусам
status_summary = (
    df_o['order_status']
    .value_counts()
    .to_frame(name='order_count')
    .reset_index()
    .rename(columns={'index': 'order_status'})
)

status_summary['share_pct'] = (
    status_summary['order_count']
    / status_summary['order_count'].sum()
    * 100
)

status_summary = status_summary.sort_values('order_count', ascending=False)

status_summary
```

```python
# посмотрим пустые значения в датах
missing_dates = (
    df_o[date_cols]
    .isna()
    .sum()
    .to_frame(name='missing_count')
)

missing_dates['missing_share_pct'] = (
    missing_dates['missing_count']
    / len(df_o)
    * 100
).round(2)

missing_dates = missing_dates.sort_values('missing_share_pct', ascending=False)

missing_dates
```

```python
# проверяем в каких статусах отсутствуют значения в колонке approved_at
mask = df_o['approved_at_date'].isna()

df_o.loc[mask, 'order_status'].value_counts()

# видим, что у нас имеется 9 доставленных заказов, которые не прошли этап согласования
```

```python
# проверяем в каких статусах отсутствуют значения в колонке delivered
mask = df_o['delivered_date'].isna()

df_o.loc[mask, 'order_status'].value_counts()

# в основном пропуски логичные, кроме 6 заказов delivered.
# заказы имеют статус delivered, но при этом отсутствуют даты в столбце delivered
```

```python
df_o.columns
```

```python
# рассчитаем длительность этапов заказа
df_o['dt_purchased_approved'] = (df_o['approved_at_date'] - df_o['purchased_date']).dt.days.astype('float')
df_o['dt_approved_delivered'] = (df_o['delivered_date'] - df_o['approved_at_date']).dt.days.astype('float')
df_o['dt_purchased_estimated'] = (df_o['estimated_delivery_date'] - df_o['purchased_date']).dt.days.astype('float')

df_o.head()
```

```python
df_o.dtypes
```

```python
# посмотрим основные статистики для разности дат, чтобы выявить аномалии
df_o[['dt_purchased_approved', 'dt_approved_delivered',
       'dt_purchased_estimated']].describe()

# dt_purchased_approved
# минимальное и среднее значение равны нулю => заказы одобряются день-в-день
# но максимальное значение равно 60 дням => редкий кейс или ошибка даты?

# dt_approved_delivered
# минимальное значение отрицательное => доставка раньше, чем одобрение
# максимальное значение 208 дней => отдаленный штат, ошибка даты или сильно проблемный заказ?

# dt_purchased_estimated
# максимальный обещанный срок 149 дней => редкая позиция?
# наверняка этот заказ и доставлялся дольше всего
```

```python
# рассмотрим аномалии в датах подробнее
df_o[df_o['dt_purchased_approved'] > 30]
```

```python
df_o[df_o['dt_approved_delivered'] > 60].head()
```

```python
df_o[df_o['dt_purchased_estimated'] > 60].head()
```

```python
# пометим флагами аномальные даты (отрицательные и с задержками)
# задержкой будем считать разницу дат:
# dt_purchased_approved > 30 дней
# dt_approved_delivered > 60 дней
# dt_purchased_estimated > 90 дней
df_o['is_negative_delivery_time'] = df_o['dt_approved_delivered'] < 0
df_o[df_o['is_negative_delivery_time'] == True].head()
```

```python
df_o[df_o['is_negative_delivery_time'] == True].is_negative_delivery_time.count()
```

```python
df_o['is_too_long_delivery_time'] = df_o['dt_approved_delivered'] > 90
df_o[df_o['is_too_long_delivery_time'] == True].head()
```

```python
df_o[df_o['is_too_long_delivery_time'] == True].is_too_long_delivery_time.count()
```

```python
df_o['is_too_long_approval_time'] = df_o['dt_purchased_approved'] > 30
df_o[df_o['is_too_long_approval_time'] == True]
```

```python
df_o[df_o['is_too_long_approval_time'] == True].is_too_long_approval_time.count()
```

```python
df_o['is_too_long_estimated'] = df_o['dt_purchased_estimated'] > 90
df_o[df_o['is_too_long_estimated'] == True].head()
```

```python
df_o[df_o['is_too_long_estimated'] == True].is_too_long_estimated.count()
```

```python
# соберем всю информацию в одну табицу и посмотрим количество аномальных дат
date_flag_cols = ['is_negative_delivery_time',
       'is_too_long_delivery_time', 'is_too_long_approval_time',
       'is_too_long_estimated']

flag_counts = (df_o[date_flag_cols]
               .sum()
               .to_frame(name='flag_counts'))

flag_counts['flag_share_pct'] = (
    flag_counts['flag_counts']
    / len(df_o)
    * 100
    ).round(2)

flag_counts = flag_counts.sort_values('flag_counts', ascending=False)

flag_counts

# невысокие показатели аномальных дат позволяют их исключить из дальнейшего анализа в виду невозможности уточнить у бизнеса
# будь возможность уточнить у бизнеса, то первым делом я бы уточнила "негативные даты", вероятно случилась ошибка при заполнении и даты перепутали в соседних колонках
# по датам с длинной доставкой я бы посмотрела, что за продукт, куда доставлялся
```

```python
# пометим аномальные заказы флагом и исключим их из дальнейшего анализа
df_o['is_anomaly'] = (
    df_o['is_negative_delivery_time']
    | df_o['is_too_long_delivery_time']
    | df_o['is_too_long_approval_time']
    | df_o['is_too_long_estimated']
)

df_o_clean = df_o[~df_o['is_anomaly']]
df_o_clean.head()
```

```python
# оценим динамику заказов по дням / месяцам
df_o_clean['date'] = df_o_clean['purchased_date'].dt.date
df_o_clean['year_month'] = df_o_clean['purchased_date'].dt.to_period('M')
df_o_clean['year'] = df_o_clean['purchased_date'].dt.year
df_o_clean['month'] = df_o_clean['purchased_date'].dt.month
df_o_clean['day'] = df_o_clean['purchased_date'].dt.day
df_o_clean['week'] = df_o_clean['purchased_date'].dt.isocalendar().week
df_o_clean['day_of_week'] = df_o_clean['purchased_date'].dt.day_name()
df_o_clean['hour'] = df_o_clean['purchased_date'].dt.hour

df_o_clean.head()
```

```python
df_o_clean['purchased_date'].describe()
```

```python
daily = df_o_clean.groupby('date').size()

plt.figure(figsize=(15, 6))
daily.plot(kind='line', linewidth=1)
plt.title('Количество заказов по дням', fontsize=14)
plt.ylabel('Количество заказов')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
print(f"Всего дней: {len(daily)}")
print(f"Среднее заказов в день: {daily.mean():.2f}")
print(f"Медиана: {daily.median():.2f}")
print(f"Максимум: {daily.max()} (в {daily.idxmax()})")
print(f"Минимум: {daily.min()} (в {daily.idxmin()})")
```

```python
year_month = df_o_clean.groupby('year_month').size()

plt.figure(figsize=(15, 6))
year_month.plot(kind='bar', width=0.8, color='steelblue', linewidth=0.5)
plt.title('Количество заказов по месяцам', fontsize=14)
plt.ylabel('Количество заказов')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
print(f"Всего месяцев: {len(year_month)}")
print(f"Среднее заказов в месяц: {year_month.mean():.2f}")
print(f"Медиана: {year_month.median():.2f}")
print(f"Максимум: {year_month.max()} (в {year_month.idxmax()})")
print(f"Минимум: {year_month.min()} (в {year_month.idxmin()})")
```

```python
monthly_by_year = df_o_clean.groupby(['year', 'month']).size().reset_index(name='order_id')
monthly_by_year = monthly_by_year[monthly_by_year['year'].isin([2017, 2018])]

years = sorted(monthly_by_year['year'].unique())

plt.figure(figsize=(15, 6))

for year in years:
    data = monthly_by_year[monthly_by_year['year'] == year]
    plt.plot(data['month'], data['order_id'], 'o-', linewidth=2, markersize=6, label=f'{year}')

plt.title('Помесячные продажи по годам', fontsize=14)
plt.ylabel('Количество заказов')
plt.xticks(range(1, 13), ['Янв', 'Фев', 'Мар', 'Апр', 'Май', 'Июн',
                          'Июл', 'Авг', 'Сен', 'Окт', 'Ноя', 'Дек'])
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
day_of_week = df_o_clean.groupby('day_of_week').size()

ru_days = {
    'Monday': 'Понедельник',
    'Tuesday': 'Вторник',
    'Wednesday': 'Среда',
    'Thursday': 'Четверг',
    'Friday': 'Пятница',
    'Saturday': 'Суббота',
    'Sunday': 'Воскресенье'
}

en_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
day_of_week = day_of_week.reindex(en_order)

ru_labels = [ru_days[day] for day in day_of_week.index]

plt.figure(figsize=(15, 6))
plt.bar(range(len(day_of_week)), day_of_week.values, color='steelblue')
plt.title('Количество заказов по дням недели', fontsize=14)
plt.ylabel('Количество заказов')
plt.xticks(range(len(day_of_week)), ru_labels, rotation=0, ha='right')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
print(f"Всего заказов: {day_of_week.sum()}")
print(f"Среднее в день: {day_of_week.mean():.2f}")
print(f"Самый загруженный день: {ru_days[day_of_week.idxmax()]} ({day_of_week.max()} заказов)")
print(f"Самый спокойный день: {ru_days[day_of_week.idxmin()]} ({day_of_week.min()} заказов)")
```

```python
hourly = df_o_clean.groupby('hour').size()
hourly = hourly.reindex(range(24), fill_value=0)

plt.figure(figsize=(15, 6))
plt.bar(hourly.index, hourly.values, color='steelblue', width=0.8)
plt.title('Количество заказов по часам', fontsize=14)
plt.xlabel('Час дня')
plt.ylabel('Количество заказов')
plt.xticks(range(0, 24, 2), [f'{h}:00' for h in range(0, 24, 2)])
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
top_hours = sorted(hourly.nlargest(3).index)
print(f"Пиковые часы: {', '.join([f'{h}:00' for h in top_hours])}")
```
